## Supabase connection test

```python
load_dotenv(dotenv_path="../.env")

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")

print(f"URL loaded: {bool(SUPABASE_URL)}")
print(f"KEY loaded: {bool(SUPABASE_KEY)}")
```

```python
supabase = create_client(SUPABASE_URL, SUPABASE_KEY)
```

```python
response = supabase.table("sales").select("*").execute()
```

```python
df = pd.DataFrame(response.data)
df.head()
```

```python
df.info()
```

## Basic sales analysis

```python
df["sale_date"] = pd.to_datetime(df["sale_date"])
```

```python
daily_sales = df.groupby("sale_date")["total_price"].sum().reset_index()
```

```python
daily_sales.head()
```

```python
fig = px.line(
    daily_sales,
    x="sale_date",
    y="total_price",
    title="Müük ajas"
)
fig.show()
```

```python
df["year_month"] = df["sale_date"].dt.to_period("M").astype(str)
monthly_sales = df.groupby("year_month")["total_price"].sum().reset_index()
```

```python
monthly_sales.head()
```

```python
fig = px.bar(
    monthly_sales,
    x="year_month",
    y="total_price",
    title="Kuupõhine müük"
)
fig.show()
```
